```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import pandas as pd

url = 'https://www4.stat.ncsu.edu/~boos/var.select/diabetes.tab.txt'
data = pd.read_csv(url, sep='\t')
print(data.head())
```

# Task 2: Plotting boxplots for BMI, BP and Y depending on gender

```python
for col in ['BMI', 'BP', 'Y']:
  data.boxplot(column=col, by='SEX')
  plt.show()
```

# Task 3: Question? - What is the the distribution of Age, Sex, BMI and Y variables?

```python
for col in ['AGE', 'SEX', 'BMI', 'Y']:
  data[col].hist()
  plt.show()
```

# Task 4: Testing the correlation between different variables and disease progression (Y)

```python
data.corr()
```

```python
# Calculate the correlation matrix
corr_matrix = data.corr()

# heatmap
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

## Observation:
* The strongest correlation of Y is BMI and S5(blood sugar)

```python
fig, ax = plt.subplots(1,3,figsize=(10,5))
for i,n in enumerate(['BMI','S5','BP']):
    ax[i].scatter(data['Y'],data[n])
    ax[i].set_title(n)
plt.show()
```

# Task 5: Testing the hypothesis that the degree of diabetes progression is different between men and women

```python
from scipy.stats import ttest_ind

tval, pval = ttest_ind(data.loc[data['SEX']==1,['Y']], data.loc[data['SEX']==2,['Y']],equal_var=False)
print(f"T-value = {tval[0]:.2f}\nP-value: {pval[0]}")
```
